This is the IMDB 250 Clone. It is a simple recommender system that recommends movie based on a single metric which is the weighted rating. It is a purely metric based recommender system. This work was done by following the tutorial from the book "Hands on recommender systems" by PacktPublishing.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('movies_metadata.csv')
df.head()
```

```python
#Number of votes garnered by 90th percentile 
m = df['vote_count'].quantile(0.90)
m
```

```python
#Considering movies longer than 45 minutes and less than 300 minutes 
q_movies = df[(df['runtime'] >= 45) & (df['runtime'] <= 300)]

#Consider movies that garnered more than m votes 
q_movies = q_movies[q_movies['vote_count'] >= m]
q_movies 
```

```python
#Calculate mean of average votes column 
C = df['vote_average'].mean()
C
```

```python
#Function to calculate weighted rating 
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    #Compute weighted score 
    return (v/(v+m)*R) + (m/(m+v)*C)
```

```python
q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
```

```python
q_movies = q_movies.sort_values('score', ascending=False)
q_movies[['title', 'vote_count', 'vote_average', 'score', 'runtime']].head(10)
```
